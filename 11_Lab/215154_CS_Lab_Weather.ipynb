<h2>Missing Values</h2>

<h4>. Checking Null Values for Int and Float data type columns</h4>

```python
nan_int_float_indices= {}

int_float_columns = merged_dataset.select_dtypes(include=['int64','float64'])

for column in int_float_columns.columns:
    mask= int_float_columns[column].isna()  # filter the rows with NaN 
    indices = int_float_columns[column][mask].index.tolist()  # find row index and converting to list
    nan_int_float_indices[column] = indices # storing in dictionary
    

for column, list in nan_int_float_indices.items():
     print(f"{column}: {len(list)} NaN values")  # count each list NaN


# nan_int_float_indices
```

<h4>. Checking NaN (shows None in excel) or Empty('') and Zero('0')  value in Object data type columns</h4>

```python
object_type_list = df1.select_dtypes(include=['object'])
print("NaN values :", object_type_list.isna().sum())
```

<h2>Null, Empty and None values removal</h2>

<h4>For temperature and days, taking average is best approch, becuase we can say if the value is missing then the temperature and days count will be average</h4>

```python
print (merged_dataset['Days of accumulation of maximum temperature'].mean())
print (merged_dataset['Days of accumulation of minimum temperature'].mean())
print (merged_dataset['Maximum temperature (Degree C)'].mean())
print (merged_dataset['Minimum temperature (Degree C)'].mean())
```

```python

c1='Days of accumulation of maximum temperature'
c2='Days of accumulation of minimum temperature'
c3='Maximum temperature (Degree C)'
c4='Minimum temperature (Degree C)'
c5='Quality'

m1= merged_dataset[c1].mean()
m2= merged_dataset[c2].mean()
m3= merged_dataset[c3].mean()
m4= merged_dataset[c4].mean()
m5=pd.to_numeric(merged_dataset[c5].mean())




merged_dataset[c1]=merged_dataset[c1].fillna(m1)
merged_dataset[c2]=merged_dataset[c2].fillna(m2)
merged_dataset[c3]=merged_dataset[c3].fillna(m3)
merged_dataset[c4]=merged_dataset[c4].fillna(m4)

merged_dataset[c5]=pd.to_numeric(merged_dataset[c5].fillna(m5))




merged_dataset.head()

```
